```python
from keras._tf_keras.keras.models import Sequential
from keras._tf_keras.keras.layers import Conv2D, MaxPooling2D, Flatten, BatchNormalization, Activation, Input, Dropout, Dense
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
x_train_batches = []
y_train_batches = []

for i in range (1,6):
    batch = unpickle('Data/cifar-10/data_batch_1')

    x_train_batches.append(batch[b'data'])
    y_train_batches.append(batch[b'labels'])

x_train = np.concatenate(x_train_batches, axis=0)
y_train = np.concatenate(y_train_batches, axis=0)
```

```python
test_batch = unpickle('Data/cifar-10/test_batch')

x_test = test_batch[b'data']
y_test = test_batch[b'labels']
```

```python
x_train = x_train.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)
x_test = x_test.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)

y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
cifar10_classes = [
    "airplane", "automobile", "bird", "cat", "deer", 
    "dog", "frog", "horse", "ship", "truck"
]

numberOfClasses = len(cifar10_classes)
```

```python
model = Sequential([
    Input(shape=(32, 32, 3)),
    Conv2D(filters=96, kernel_size= (3, 3), padding = 'same'),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.2),
    Conv2D(filters=96, kernel_size= (3, 3), padding = 'same'),
    Activation('relu'),
    Conv2D(filters=96, kernel_size= (3, 3), padding = 'same'),
    Activation('relu'),
    MaxPooling2D(pool_size= (3,3), strides= 2),
    Dropout(0.2),

    Conv2D(filters=192, kernel_size= (3, 3), padding = 'same'),
    BatchNormalization(),
    Activation('relu'),
    Conv2D(filters=192, kernel_size= (3, 3), padding = 'valid'),
    Activation('relu'),
    Conv2D(filters=192, kernel_size= (3, 3), padding = 'same'),
    Activation('relu'),
    MaxPooling2D(pool_size= (3,3), strides= 2),
    Dropout(0.2),

    Conv2D(filters=192, kernel_size= (3, 3), padding = 'same'),
    BatchNormalization(),
    Activation('relu'),
    Conv2D(filters=192, kernel_size= (1, 1), padding = 'valid'),
    Activation('relu'),
    Conv2D(filters=10, kernel_size= (1, 1), padding = 'valid'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(),

    Flatten(),
    Dense(units=1024),
    Activation('relu'),
    Dropout(0.5),
    Dense(units=numberOfClasses),
    Activation('softmax')
])
```
